```python
dict1=dict(visa_df['continent'].value_counts())
Keys=[]
values=[]
for i in dict1:
    Keys.append(i)
    values.append(dict1[i])
continent_df=pd.DataFrame(zip(Keys,values),columns=['Continent','Count'])
continent_df.to_csv('/Users/srilakshmidurgachunduri/Documents/NareshIT/Data Science/EDA/Data Files/continent_df.csv',index='false')
continent_df
```
