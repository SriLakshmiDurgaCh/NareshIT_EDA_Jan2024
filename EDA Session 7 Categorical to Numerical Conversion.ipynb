```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

visa_dataset="/Users/srilakshmidurgachunduri/Documents/NareshIT/Data Science/EDA/Data Files/Visadataset.csv"
visa_df=pd.read_csv(visa_dataset)
visa_df.head()
```

$Map Method$

```python
- first consider one categorical column

- Observe how many unique values are there

- For example consider case status is visa_df

- unique values are "Certified" and "Denied"

- Create a dictionary using Key:Value pairs
    
    - d1={'Certified':0,'Denied':1}
```

```python
visa_df['case_status'].unique()
```

```python
d1={'Certified':0,'Denied':1}
visa_df['case_status']=visa_df['case_status'].map(d1)
visa_df
```

```python
visa_df['case_status']
```

```python
visa_df['continent'].unique()
```

```python
d2={'Asia':0, 'Africa':1, 'North America':2, 'Europe':3, 'South America':4,
       'Oceania':5}
visa_df['continent']=visa_df['continent'].map(d2)
visa_df
```

```python
# instead of mapping numerical values mannually we can try different methods
```

```python
d2={}
labels=visa_df['continent'].unique()
for i in range(len(labels)):
    d2[labels[i]]=i

visa_df['continent']=visa_df['continent'].map(d2)
visa_df
```

**Method 3:**
    
$One-hot-encoder$

pd.getdummies()



- one hot encoder(OHE) means at a time only one will be ON and other will be OFF

- ON represents with 1 and OFF represents with 0

- One hot encoder creates new extra columns based on labels in the column

- for ex: column case status has 2 unique labels
    
- If we create one hot encoder on case_status 2 new coulumns are created, they are case_status_Certified, case_status_denied

|case_status|case_status_Certified|case_status_Denied|
|-----------|---------------------|------------------|
|Denied|      1                   | 0                |
|Certified| 0                     |   1              |

**Advantage**

- OHE gives independency between the variables 

- It is one of the imp property before applying ML models

- variables are independent each other

- variables are perpendicular each other

- variables are Orthogonal to each other

- variables are 90° phase shift.

**Disadvantages**

- assume that a column has 100 new columns ,

- If we apply OHE it creates 100 new columns and 

- Your data matrix will be sparse. (more columns)

- We require more time more memory to compute calculations.

- This is called "Curse of Dimensionality"

**Method 4:**
    
$ Label-Encoder $

- we already implemented map method by our own

- The same map method implemented as Python package way

- Package name: sklearn

- Method name: LabelEncoder

- Sklearn package approach is very easy

    - Step-1: Read the package
    
    - Step-2: Save the package
    
    Read data again

    - Step-3: Apply fit transform

```python
visa_dataset="/Users/srilakshmidurgachunduri/Documents/NareshIT/Data Science/EDA/Data Files/Visadataset.csv"
visa_df=pd.read_csv(visa_dataset)

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

visa_df['continent']=le.fit_transform(visa_df['continent'])
visa_df
```

```python
visa_dataset="/Users/srilakshmidurgachunduri/Documents/NareshIT/Data Science/EDA/Data Files/Visadataset.csv"
visa_df=pd.read_csv(visa_dataset)
cat_columns=visa_df.select_dtypes(include='object').columns
for i in cat_columns:
    visa_df[i]=le.fit_transform(visa_df[i])

visa_df.head()    

```
